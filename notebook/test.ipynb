```python
import scipy
import scipy.stats
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.linspace(-1, 1, 100)
```

```python
# dist = scipy.stats.beta
# y = dist.pdf(x, 1+23, 50-24)

dist = scipy.stats.norm(0,0.1)
y = dist.pdf(x)
```

```python
plt.plot(x, y)
plt.show()
```
